Imports

```python
#imports for neural network
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split
from torchsummary import summary

#imports for vision
import torchvision
import torchvision.transforms as transforms
from torchvision import datasets
from torchvision.datasets import ImageFolder
from torchvision.utils import make_grid


#imports for visualizations
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

from PIL import Image
```

Raw Image

```python
#processing path
pistol_path = 'demo_images/pistol_demo.jpg'
no_pistol_path = 'demo_images/no_pistol_demo.jpg'

pistol_image = Image.open(pistol_path)
no_pistol_image = Image.open(no_pistol_path)

#plotting first
plt.subplot(1, 2, 1)
plt.imshow(pistol_image)
plt.title('Raw Concealed Pistol Image')
plt.axis('off')

#plotting second
plt.subplot(1, 2, 2)
plt.imshow(no_pistol_image)
plt.title('Raw No Pistol Image')
plt.axis('off')

#showing images
plt.tight_layout()
plt.show()
```

Preprocessed Image

```python
transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.Grayscale(),
    transforms.ToTensor()
])

#transforming image
pistol_image = transform(Image.open(pistol_path))
no_pistol_image = transform(Image.open(no_pistol_path))

#resizing to be able to be plotted
pistol_image = pistol_image.squeeze().numpy()
no_pistol_image = no_pistol_image.squeeze().numpy()


#plot first
plt.subplot(1, 2, 1)
plt.imshow(pistol_image, cmap = 'gray')
plt.title('Processed Concealed Pistol Image')
plt.axis('off')

#plot second
plt.subplot(1, 2, 2)
plt.imshow(no_pistol_image, cmap = 'gray')
plt.title('Processed No Pistol Image')
plt.axis('off')

#showing plot
plt.tight_layout()
plt.show()
```
